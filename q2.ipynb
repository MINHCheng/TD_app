```python
import pandas as pd
from pandas import Timestamp as Timestamp
import plotly.express as px
```

```python
df = pd.read_csv('./extracted/data/q2/filtered_data.csv')
```

```python
crisis = {
    'GFC' : (Timestamp(2007, 1, 1 ) ,Timestamp( 2009 , 12 , 31 ) ) , # G l o b a l F i n a n c i a l
    'EMC' : (Timestamp( 1997, 1, 1 ) ,Timestamp( 1997 , 12 , 31 ) ) , # Asian Emerging M
    'LTCM' : (Timestamp( 1998, 1 ,1 ) ,Timestamp( 1998 , 12 , 31 ) ) , # Long=Term C a p i t
    'Dotcom' : (Timestamp ( 2000, 1 ,1 ) ,Timestamp( 2002 , 12 , 31 ) ) , # Dot=com B u b b l
    'ARG' : (Timestamp( 2001, 1 ,1 ) ,Timestamp( 2001 , 12 , 31 ) ) , # A r g e n t i n a D e f a u l
    'EUR' : (Timestamp( 2010,1 ,1 ) ,Timestamp(2012 , 12 , 31 ) ) , # Eurozone S o v e r e i g
    'COVID' : (Timestamp( 2020, 3 ,1 ) ,Timestamp( 2020 , 3, 31 ) ) , # Pandemic=Induce
    'UK' : (Timestamp( 2022, 9 ,1 ) ,Timestamp( 2022 , 10 , 31 ) ) , # UK G i l t C r i s i s
    'SVB' : (Timestamp( 2023, 3 ,1 ) ,Timestamp( 2023 ,3 , 31 ) ) # S i l i c o n V a l l e y Ban
}
```

```python
## decoration/wrapper for the base function -- mainly for the crisis case
def new_plot(crisis):
    def wrapper(func):
        def inner(*args):
            signal = args[0]
            fig = func(*args)

            if signal == 'Plot with Crisis Periods':
                for key, (start, end) in crisis.items():
                    fig.add_vrect(
                        x0=start, x1=end,
                        fillcolor="red", opacity=0.2,
                        layer="below", line_width=0,
                        annotation_text=key, 
                        annotation_position="top left"
                    )
            
            fig.show()
            return fig
        return inner
    return wrapper 
```

```python
# base plot using plotly library
@new_plot(crisis)
def base_plot(signals:str):
    fig = px.line(df, x="Date", y="Value", title=signals, template="plotly_dark")
    return fig
```

```python
mplot = base_plot("Main Plot")
mplot = base_plot("Plot with Crisis Periods")\
```
